```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img=cv2.imread('ss.png',0)  #read in grey scale  #jidha grey component zyada wahai brighter
plt.imshow(img)
plt.imshow(img,cmap= 'gray')




```

```python
ret,threh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)  #value above this 1   below 0
ret  #actual cutoff value 
plt.imshow(threh1,cmap='gray')

#more functions thresholf_biary_inv
#thresholf trunc based on function

```
