```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def display(img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')

reeses=cv2.imread('cf.jpg',0)
reeses=cv2.resize(reeses,(400,400))
display(reeses)
```

```python
cereals=cv2.imread('inshelf.webp',0)
display(cereals)
```

```python
orb=cv2.ORB_create()
```

```python
kp1,des1=orb.detectAndCompute(reeses,None)  # add features of the image
#larger 
kp2,des2=orb.detectAndCompute(cereals,None)  # add features of the image


```

```python
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
matches=bf.match(des1,des2)

#sort inorder of itances 
matches=sorted(matches,key=lambda x:x.distance)
reeses_matches=cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2) #25 linking lines
```

```python
display(reeses_matches)   #this is not that accurate 
```
