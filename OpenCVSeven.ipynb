```python
#stacking images by a function
#new pixel= a*pixel_1+b*pixel_2+gamma
#first pixel dominate->higher a 
import cv2
img1=cv2.imread('s.jpeg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('ss.png')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(img1)
plt.imshow(img2)
```

```python
#image blending
img1=cv2.resize(img1,(1200,1200))
img2=cv2.resize(img2,(1200,1200))
plt.imshow(img1)
plt.imshow(img2)

blended=cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
#overlay smaller on top of othe
img22=cv2.resize(img2,(2400,2400))
# add img1 over img22
x_offset=0
y_offset=0

t=img1.shape
x_end=x_offset+t[1]
y_end=y_offset+t[0]

img22[y_offset:y_end,x_offset:x_end]=img1
plt.imshow(img22)
```

```python
#grab small part og thr image 
rows,cols,channels=img22.shape
print(rows)
print(cols)
print(channels)

roi=img22[y_offset:100,x_offset:100]
plt.imshow(roi)

```

```python
mask_inv=cv2.bitwise_not(img22)
plt.imshow(mask_inv)
```
