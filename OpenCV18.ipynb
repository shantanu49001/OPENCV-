```python
#corner detection
#corner os a pt whose neigbour stands at two dimnant directons or thr pt wher e thr constrast changes 
#signofcnt hanges in all directions

#flat regions-->no change in any direction

#shi tamoshi corner detection:=>
#harris corner detection

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess=cv2.imread('chessb.jpg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess,cmap='gray')
```

```python
real_chess=cv2.imread('r_chess.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess,cmap='gray')
```

```python
gray=np.float32(gray_flat_chess)  #imp to change to flsast
dst=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k= 0.04 ) #k defaylt 0.04 move up if required -->harris detctor  free constsnt 
dst=cv2.dilate(dst,None)
flat_chess[dst> 0.01*dst.max()]=[255,0,0] #jifha bhi dtet ho
plt.imshow(flat_chess)
```

```python
#testing on real image  detcts way more edges 
gray=np.float32(gray_flat_chess)
dst=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dst=cv2.dilate(dst,None)
#both must have same size 
real_chess[dst> 0.01*dst.max()]=[255,0,0]
plt.imshow(real_chess)

```

```python
conrers=cv2.goodFeaturesToTrack(gray_flat_chess,maxCorners=57,qualityLevel=0.01,minDistance=10)  #qulaity -->minimum eighen value
conrers=np.int0(conrers)
for i in conrers:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```
