```python
#placing the grid like pattern over a object 
#as grid move the object move
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess=cv2.imread('chessb.jpg')
plt.imshow(flat_chess)
```

```python
found,corners=cv2.findChessboardCorners(flat_chess,(7,7))  # last wala detct nhi kr payega specifically  foe chessboard like inputs
found
```

```python
corners
```

```python
cv2.drawChessboardCorners(flat_chess,(7,7),corners,found)
plt.imshow(flat_chess)
```
