```python
# face detection
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline

#denis nadia solvay



```

```python
nadia=cv2.imread('nadia.jpg',0)  #gray scaled import
#nadia.resize((600,400))
denis=cv2.imread('denis.jpg',0)
solvay=cv2.imread('solvay.jpg',0)

```

```python
plt.imshow(solvay,cmap='gray')
```

```python
#data folder has xml file for the features 
#pre rained xml file
face_cascade=cv2.CascadeClassifier('DATA\haarcascades\haarcascade_frontalface_default.xml')

```

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)  #min neigb only for image wit muitle faces 

    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)

        return face_img
```

```python
result=detect_face(solvay)
plt.imshow(result,cmap='gray')
```

```python
#live video se face detet -->just video=frames images 
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read(0)

    frame=detect_face(frame)
    cv2.imshow('Video ',frame)

    k=cv2.waitKey(1)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()    
```
