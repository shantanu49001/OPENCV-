```python
#morphologocal operators -->specialised kernels for specialised effects
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def load_img():
    blank_img=np.zeros((600,600))
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text= 'ABCDE',org=(50,300),fontFace=font,fontScale=5,color=(255,255,255))
    return blank_img

def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap= 'gray')

img=load_img()
display_img(img)



```

```python
kernel=np.ones((5,5),dtype=np.uint8)
result=cv2.erode(img,kernel,iterations=1)  #text ko erode krta hai
display_img(result)

```

```python
img=load_img()
white_noise=np.random.randint(low=0,high=2,size=(600,600))
display_img(white_noise)
```

```python
#put this nose to original image 
img.max()
```

```python
white_noise=white_noise*255
noise_img=white_noise+img
display(noise_img)
display_img(noise_img)
```

```python
#upper image ,e noise ke saath text ala image add ho jayega 

#but morphologic fuction se noise add ho ga but text cleaner hoga 
opening=cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
display_img(opening)
```
