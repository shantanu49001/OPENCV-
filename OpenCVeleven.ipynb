```python
#detects objeys moving 
#edge detection --> it monitors intensirt of direction og color change 

#sobel feldman oertions
#normalied x gradient me verticak edges 
#nomalised y me horxzontal edges

'''
matrix gx=[[1,0,-1]; [2,0,-2],[1,0,-1]]
gy=[[1,2,1];[0,0,0];[-1,-2,-1]]
'''

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    
```

```python
img=cv2.imread('cat.jpg',0)
plt.imshow(img)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # 1->gradient in x 0->gradienin y
display_img(sobelx)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=7) #ksize must be odd 
display_img(sobely)
```

```python
laplacian=cv2.Laplacian(img,cv2.CV_64F)  #in both direction edge detct
display_img(laplacian)
```

```python
#now we can assign propirty hoixtal me kitne edge weight dtect 
bleneded= cv2.addWeighted(src1=sobelx,src2=sobely,alpha=0.5,beta=0.5,gamma=0)
display_img(bleneded)
```
