```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('cat.jpg',0)
img.shape

plt.imshow(img,cmap= 'gray')

```

```python
image,counors,heirarchy=cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)  #    extarct both internal ans external conturs 
type(counors)

external_countors=np.zeros(np.shape)  #black 

#now rwaing the contours 
for i in range(len(counors)):
    #externa -1
    if hierarch[0][i][3]== -1:
        cv2.drawContours(external_countors,counors,i,255,-1)
        




```
