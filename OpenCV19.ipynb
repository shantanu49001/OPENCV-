```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('dogface.webp')  #not concerned about channels 
plt.imshow(img)
```

```python
edges=cv2.Canny(image=img,threshold1=50,threshold2=50) #start by assigning them mid value of the image size we have to opimum ly change them 
plt.imshow(edges)
```

```python
#best formula to judge the best thresholds 
med_value=np.median(img)
med_value # 221

lower=int(max(0,0.7* med_value))
upper=int(min(255,1.3*med_value))

edges=cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img=cv2.blur(img,ksize=(5,5))
edges=cv2.Canny(image=blurred_img,threshold1=50,threshold2=50)
plt.imshow(edges)
```
